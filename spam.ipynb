```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv("spam_or_not_spam.csv")
```

```python
dataset
```

```python
dataframe=pd.DataFrame(dataset)
```

```python
dataframe.isnull().sum()
```

```python
dataframe.fillna("",inplace=True)
```

```python
dataframe.isnull().sum()
```

```python
np.shape(dataframe)
```

```python
dataframe.info()
```

```python
X=dataframe['email']
```

```python
y=dataframe['label']
```

```python
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=6)
```

```python
model=TfidfVectorizer(min_df=1,stop_words='english')
```

```python
x_train_features=model.fit_transform(X_train)
x_test_features=model.transform(x_test)
```

```python
trained_model=LogisticRegression()
trained_model.fit(x_train_features,y_train)
```

```python
predicted=trained_model.predict(x_train_features)
accuracy=accuracy_score(y_train,predicted)
```

```python
print(accuracy)
```

```python
predicted=trained_model.predict(x_test_features)
accuracys=accuracy_score(y_test,predicted)
```

```python
print(accuracys)
```
